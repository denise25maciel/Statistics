DISTRIBUIÇÃO NORMAL(CURVA DE BELL) E SUAS PROPRIEDADES

Nesse experimento foram gerados dados aleatórios para apresentar o resultado da distribuição normal. Importante citar que, independente da distribuição dos dados (contínua, bernouli, entre outras), os mesmos apresentarão distribuição normal para amostras maiores do que o tamanho 30. Por conta desse fato, conhecer as propriedades da distribuição normal, faz-se tão importante.

```python
import plotly.figure_factory as ff
import plotly.graph_objects as go
import statistics
import random
#Creating a list of sum of 2 dice, rolled 1000 times
dados_resultado = []
for i in range(0, 1000):
    dados1 = random.randint(1, 6)
    dados2 = random.randint(1, 6)
    dados_resultado.append(dados1 + dados2)
#Calculo media, desvio padrão, mediana e moda
media = sum(dados_resultado) / len(dados_resultado)
desv_padrao = statistics.stdev(dados_resultado)
mediana = statistics.median(dados_resultado)
moda = statistics.mode(dados_resultado)
#Encontrando o posicionamento de um desvio padrão antes e depois
first_std_deviation_start, first_std_deviation_end = media-desv_padrao, media+desv_padrao
second_std_deviation_start, second_std_deviation_end = media-(2*desv_padrao), media+(2*desv_padrao)
third_std_deviation_start, third_std_deviation_end = media-(3*desv_padrao), media+(3*desv_padrao)
#Plotando o gráfico e os desvios padrão
fig = ff.create_distplot([dados_resultado], ["RESULTADO"], show_hist=False)
fig.add_trace(go.Scatter(x=[media, media], y=[0, 0.17], mode="lines", name="MEDIA"))
fig.add_trace(go.Scatter(x=[first_std_deviation_start, first_std_deviation_start], y=[0, 0.17], mode="lines", name="DESVIO PADRÃO 1"))
fig.add_trace(go.Scatter(x=[first_std_deviation_end, first_std_deviation_end], y=[0, 0.17], mode="lines", name="DESVIO PADRÃO 1"))
fig.add_trace(go.Scatter(x=[second_std_deviation_start, second_std_deviation_start], y=[0, 0.17], mode="lines", name="DESVIO PADRÃO 2"))
fig.add_trace(go.Scatter(x=[second_std_deviation_end, second_std_deviation_end], y=[0, 0.17], mode="lines", name="DESVIO PADRÃO 2"))
fig.show()
#Printando o resultado
dados1dp = [result for result in dados_resultado if result > first_std_deviation_start and result < first_std_deviation_end]
dados2dp = [result for result in dados_resultado if result > second_std_deviation_start and result < second_std_deviation_end]
dados3dp = [result for result in dados_resultado if result > third_std_deviation_start and result < third_std_deviation_end]
print("Media:{}".format(media))
print("Mediana:{}".format(mediana))
print("Moda:{}".format(moda))
print("Desvio Padrão:{}".format(desv_padrao))
print("{}% de dados em um desvio padrão".format(len(dados1dp)*100.0/len(dados_resultado)))
print("{}% de dados em dois desvios padrão".format(len(dados2dp)*100.0/len(dados_resultado)))
print("{}% de dados em três desvios padrão".format(len(dados3dp)*100.0/len(dados_resultado)))
```
